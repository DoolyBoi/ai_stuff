```python
import gymnasium as gym
from stable_baselines3 import A2C, PPO
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.callbacks import CheckpointCallback, EvalCallback
import os
```

```python
env = make_vec_env("CarRacing-v2", n_envs=4)
```

```python
env = VecFrameStack(env, n_stack=4)
```

```python
base_dir = '/mnt/c/users/haval/OneDrive - Oregon Institute of Technology/Documents/CST407/project/car_racing'
log_path = os.path.join(base_dir, 'LogsVec')
save_path = os.path.join(base_dir, 'Checkpoints')
```

```python
checkpoint_callback = CheckpointCallback(save_freq=100000, save_path=save_path, name_prefix='ppo_carracing', save_replay_buffer=True, save_vecnormalize=True,)
```

```python
model = PPO(
  "CnnPolicy",
  env,
  verbose=1,
  tensorboard_log=log_path,
  batch_size=128,
  ent_coef=0.01,
  learning_rate=0.0003,
  use_sde=True
)
```

```python
model.learn(total_timesteps=2000000, callback=checkpoint_callback)
```

```python
env2 = make_vec_env('CarRacing-v2', n_envs=1)
env = VecFrameStack(env, n_stack=4)
evaluate_policy(model=model, env=env2, n_eval_episodes=10, render=True)
```
