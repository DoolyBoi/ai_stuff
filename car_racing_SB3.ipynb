```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.callbacks import CheckpointCallback, EvalCallback
import os
```

```python
env = gym.make("CarRacing-v2")
```

```python
base_dir = '/mnt/c/users/haval/OneDrive - Oregon Institute of Technology/Documents/CST407/project/car_racing'
log_path = os.path.join(base_dir, 'Logs')
tensorboard_log=log_path
```

```python
checkpoint_callback = CheckpointCallback(save_freq=100000, save_path='./checkpoints/', name_prefix='ppo_carracing')
```

```python
model = PPO(
  "CnnPolicy",
  env,
  verbose=1,
  tensorboard_log=log_path,
  batch_size=128,
  ent_coef=0.01,
  learning_rate=0.0003,
  use_sde=True
)
```

```python
model.learn(total_timesteps = 2000000)
```

```python
model.save(base_dir)
```

```python
env2 = gym.make('CarRacing-v2', render_mode='human')
evaluate_policy(model=model, env=env2, n_eval_episodes=10, render=True)
```
